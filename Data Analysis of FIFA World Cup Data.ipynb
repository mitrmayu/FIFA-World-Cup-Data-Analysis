### Step 5: Analyzing Home Country Advantage

```python
# Making the requires calculations
home_wins = fifa_matches[fifa_matches['Home Team Goals']>fifa_matches['Away Team Goals']].shape[0]
away_wins = fifa_matches[fifa_matches['Away Team Goals']>fifa_matches['Home Team Goals']].shape[0]
total_matches = fifa_matches.shape[0]

# Calculating Home Team Win percentage
home_win_percentage = (home_wins / total_matches) * 100
print("Home Team Win Percentage:", round(home_win_percentage, 3),"%")

# Calculating Away Team Win percentage
away_win_percentage = (away_wins/total_matches) * 100
print("Away Team Win Percentage:", round(away_win_percentage, 3),"%")
```

### Step 6: Analyzing Total Goals Scored by Winning Teams

```python
# Extracting the relevant data
world_cup_goals = fifa_worldcups[['Year','Winner','GoalsScored']]

# Visualizing the total number of goals scored by the winning team
mpl.figure(figsize=(12,6))
sb.barplot(x='Year', y='GoalsScored', data=world_cup_goals, palette='turbo')
mpl.title('Total Goals Scored by Winning Team per World Cup',fontsize=17)
mpl.xlabel('Year', fontsize=13)
mpl.ylabel('Goals Scored', fontsize=13)
mpl.xticks(rotation=90)
mpl.show()
```

### Step 7: Analyzing the Average Number of Goals per Match

```python
# Calculating avergae number of goals per match per World Cup
goals_per_year = fifa_matches.groupby('Year').agg({'Home Team Goals':'sum', 'Away Team Goals':'sum'})
matches_per_year = fifa_matches.groupby('Year').size()
goals_per_year['Total Goals'] = goals_per_year['Home Team Goals'] + goals_per_year['Away Team Goals']
goals_per_year['Average Goals per Match'] = goals_per_year['Total Goals'] / matches_per_year

# Visualizing the average number of goals per match
mpl.figure(figsize=(12,6))
sb.lineplot(data=goals_per_year, x=goals_per_year.index, y='Average Goals per Match', marker = 'o', color = 'blue')
mpl.title('Average Goals per Match per World Cup', fontsize=17)
mpl.xlabel('Year', fontsize=13)
mpl.ylabel('Average Goals per Match', fontsize=13)
mpl.xticks(rotation=90)
mpl.show()
```

### Step 8: Analyzing World Cup Attendance

```python
# EXtracting the relevant data
world_cup_attendance = fifa_worldcups[['Year','Attendance']]

# Visualizing World Cup Attendance over the Years
mpl.figure(figsize=(15,6))
sb.barplot(data= world_cup_attendance, x='Attendance', y='Year', palette='viridis')
mpl.title('Attendance per World Cup', fontsize=17)
mpl.xlabel('Attendance', fontsize=13)
mpl.xticks(rotation = 90)
mpl.ylabels('Years', fontsize=13)
mpl.bar(True)
mpl.show()
```

### Step 9: Analysis of Top Goal Scorers

```python
# Extracting the required data for goal scored events
goal_events = fifa_players[fifa_players['Event'].str.contains('G', na=False)]
top_scorers = goal_events['Player Name'].value_counts().head(10)

# Visualizing data for Top Goal Scorers
mpl.figure(figsize=(12,8))
sb.barplot(x=top_scorers.index, y=top_scorers.values, palette='plasma')
mpl.title('Top 10 Goal Scorers in World Cup', fontsize=17)
mpl.xlabel('Player Name', fontsize=13)
mpl.ylabel('Goals Scored', fontsize=13)
mpl.xticks(rotation=45)
mpl.show()
```

### Step 10: Analyzing Defensive Strength

```python
# Extracting the relevant data
goals_confirmed = fifa_worldcups[['Year','GoalsScored','QualifiedTeams','MatchesPlayed']]
goals_confirmed['Total Goals'] = goals_confirmed['GoalsScored'] - goals_confirmed['QualifiedTeams'] - goals_confirmed['MatchesPlayed']

# Visualizing the goals conceded by Winning teams
mpl.figure(figsize=(12,6))
sb.barplot(x='Year', y='Total Goals', data=goals_confirmed, palette='viridis')
mpl.title('Goals Conceded by Winning Teams per World Cup', fontsize=17)
mpl.xlabel('Year', fontsize=13)
mpl.ylabel('Goals Conceded', fontsize=13)
mpl.xticks(rotation=45)
mpl.show()
```
